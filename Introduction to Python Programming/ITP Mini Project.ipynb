Q7. Print all shipments which are delivered within 7 days of courier Start date.

```python
from datetime import datetime
for i in a.values():
    if i["Delivery date"] == "Null":
        continue
    else:
        start = datetime.strptime(i["Start date"], "%d-%m-%Y")
        delivery = datetime.strptime(i["Delivery date"], "%d-%m-%Y")
        if (delivery - start).days <= 7:
            print(i)
```
